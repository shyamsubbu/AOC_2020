```python
from collections import Counter
import re
```

```python
with open('/Users/spudukko/efg.txt','r') as f:
    lines=f.readlines()
    no,v,s=0,0,set()
    for line in lines:
        no+=1
        p=re.compile(r'(\d+)-(\d+)\s(\S):\s([a-z]+)')
        m=p.search(line)
        sidx,eidx,char,ps=m.groups()
        c=Counter(ps)
        if char in c and c[char] in range(int(sidx),int(eidx)+1):
            v+=1
        
            
        
    
    
```
